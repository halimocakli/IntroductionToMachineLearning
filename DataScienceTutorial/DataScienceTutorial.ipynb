## MATPLOTLIB
Matplotlib, verinin bir çizimini(plot) oluşturarak görselleştirmemizi sağlayan kütüphanedir. Oluşturması en basit ve en çok kullanılan diyagramlar ise **Line Plot, Scatter Plot ve Histogram**'dır. Peki hangi görselleştime seçeneğini ne zaman kullanırsak verimli bir sonuç elde ederiz? 
* Grafiğin x ekseninde "zaman" değişkeni mevcutsa *Line Plot* kullanmak mantıklıdır.
* İki değişken/feature arasında korelasyon(correlation) mevcutsa *Scatter Plot* kullanmak mantıklıdır.
* Sayısal verilerin dağılımını görmek istiyorsak *Histogram* kullanmak mantıklıdır.
* Customization(Özelleştirme): Renkler, etiketler(labels), çizgi kalınlığı, başlık, saydamlık, ızgara(grid), şekil boyutu(figsize), eksen çizgileri ve çizgi stili(linestyle)

### LINE PLOT OLUŞTURALIM

```python
# Line Plot

# color = renk
# label = etiket
# linewidth = çizgi kalınlığı
# alpha = saydamlık
# grid = grid/ızgara
# linestyle = çizgi stili

plt.subplots(figsize=(15, 15))

data["Speed"].plot(kind = 'line', color = 'g', label = 'Speed', linewidth=1, alpha = 0.5, grid = True, linestyle = '-')
data["Defense"].plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')

plt.legend(loc='upper right')     # legend = eksen etiketlerini ve konumunu belirle
plt.xlabel('x axis')             
plt.ylabel('y axis')
plt.title('Line Plot')
plt.show()
```

### SCATTER PLOT OLUŞTURALIM

```python
# Scatter Plot

# İki feature arasındaki positive correlation ve negative correlation değişimlerini inceleyeceğiz
# Feature'lar;
    # x = Attack
    # y = Defense

data.plot(kind='scatter', x='Attack', y='Defense', alpha = 0.5, color = 'red', figsize=(15,7))
plt.xlabel('Attack')
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot')
plt.show()
```

### DICTIONARY(SÖZLÜK)
Neden dictionary veri yapısına ihtiyaç duyarız?
* Çünkü dictionary, tıpkı küçük bir dataframe gibi içerisinde 'key' ve 'values' değerlerini bulundurur.
* Dictionary, list veri yapısından daha hızlı çalışır.
<br>
'key' ve 'value' değerleri nedir? Örnek:
* dictionary = {'Spain' : 'Madrid'}
* Key >>> Spain
* Values >>> Madrid
<br>

**İşte bu kadar basit**

Şimdi de, keys(), values(), güncelleme(update), ekleme(add), kontrol etme(chech), key silme(remove key), tüm girdileri silme(remove all entries) ve dictionary silme(remove dictinary) gibi metot ve özellikler ile alakalı uygulamalar yapalım.

```python
# Bir dictionary yaratıp key ve values değerlerini inceleyim

dictInstance = {
    'Spain' : 'Madrid',
    'USA' : 'Vegas'
}

print(dictInstance.keys())
print(dictInstance.values())

# Görüldüğü üzere dictInstance adlı dictionary yapısının;

    # Key değerleri >> Spain ve USA
    # Values değerleri >> Madrid ve Vegas
```

```python
# Dictionary key'leri; tıpkı string, boolen, float, integer ya da tuple gibi immutable nesnelerdir.
# List veri yapısı mutable'dır. Bu nedenle key değeri olarak bir liste tercih edilemez.
# Key değerleri unique yani eşşizdir.

print(dictInstance)                    # dictInstance dictionary'sini yazdır ve ilk halini incele.

dictInstance['Spain'] = "Barcelona"    # Halihazırda var olan key'in values değişkenini güncelledik.
print(dictInstance)

dictInstance['France'] = "Paris"       # Yeni bir key-values ikilisini dictInstance yapımıza ekledik.
print(dictInstance)

del dictInstance['Spain']              # Spain key'ini, values değişkeni ile birlikte sildik.
print(dictInstance)

print('France' in dictInstance)        # France key'inin dictInstance yapısı içerisinde olup olmadığını kontrol et.

dictInstance.clear()                   # dictInstance içerisindeki tüm girdileri sil.
print(dictInstance)
```

```python
# Boş bir dictionary yapısının bellekte yer kaplaması mantıklı değil.
# Bu nedenle clear() metodu ile içini boşalttığımız dictInstance'ı tamamen silebiliriz.
# del dictInstance ifadesi dictionary yapısını silmemizi sağlar.
# Silinen bir dictionary'i yazdırmak hata vereceği için print() satırını yorum satırı yaptık.

del dictInstance
# print(dictInstance)      
```

### WHILE AND FOR LOOPS - WHILE VE FOR DÖNGÜLERİ
* `while` ve `for` döngülerini en basit şekilde öğreneceğiz.

```python
# (i != 5) şartı True olduğu sürece while döngüsü içerisinde kal.

i = 0
while i != 5:
    print('i is: ', i)
    i +=1
print(i,' is equal to 5')
```

```python
print("-------For List-------")

# for döngüsü örneği

lis = [1,2,3,4,5]
for i in lis:
    print(f'i is: {i}')
print('')

print("-------For Enumerate-------")

# enumerate -> liste yapısı içerisindeki değerleri ve o değerlerin listedeki indekslerini yazdırmamızı sağlar
# index : value = 0:1, 1:2, 2:3, 3:4, 4:5

for index, value in enumerate(lis):
    print(index," : ",value)
print('')   

print("-------For Dictionary-------")

# Dictionary yapısında döngüleri inceleyelim
# Dictionary'nin key ve values değerlerine erişebilmek için for döngüsü oluşturabiliriz.
# Bunu yapabilmek için dictionary veri yapısına ait olan items() metodunu kullanabiliriz.

dictionary = {
    'Spain':'Madrid',
    'France':'Paris'
}

for key, value in dictionary.items():
    print(key," : ",value)
print('')

print("-------For Pandas-------")

# Pandas verilerine erişim sağlamak için for döngüsü kullanabiliriz.
# Attack feature'unun ilk indeksindeki değeri -> [0:3] 
# iterrows() metodu ile, belirtilen indeksi ve o indeksteki değeri alıp index ve value değişkenlerine atıyoruz.
# Ardından for döngüsü ile yazdırıyoruz
for index, value in data[['Attack']][0:3].iterrows():
    print(index," : ",value)
```

### USER DEFINED FUNCTION - KULLANICI TANIMLI FONKSİYONLAR
Peki, fonksiyonlar ile ilgili bilmemiz gereken kavramlar nelerdir?
* docstrings: ait olduğu fonksiyonun işlevini özetleyen dökümanlardır. Örneğin:

def f():
    <br>"""This is docstring for documentation of function f"""


* Tuple: birbiri ardına sıralanmış Python objelerini tutan veri yapısı.
    1. Tuple içerisindeki değerler değiştirilemez çünkü tuple immutable'dır.
    2. Tuple içerisindeki sıralı objeler parantez içerisinde belirtilir -> tupleInstance = (a,b,c)
    3. Tuple içerisinde depolanan verileri birden fazla değişkene aktarabiliriz ve bu işleme `unpacking` denir

```python
# Öğrendiklerimizi uygulayalım

def tupleInstance():
    """ return defined t tuple"""  # docstring
    t = (1,2,3)
    return t

a,b,c = tupleInstance()     # unpacking
print(a,b,c)
```

### SCOPE - KAPSAM / FALİYET ALANI
Peki, scope hakkında bilmemiz gereken kavramlar nelerdir?
* local: Bir fonksiyon içerisinde tanımlanan değişkenlerin türü
* enclosed: Eğer yerelde tanımlı değilse kapsayan fonksiyon içinde yer alan değer kullanılır
* global: main body(ana gövde) içerisinde tanımlanan değişkenlerin türü
* built in scope: Bizim tanımlamadığımız, Python ile birlikte gelen fonksiyon ya da değişkenlerdir. Örn: print(), len()
<br><br>Lets make some basic examples

```python
# print() fonksiyonu sonucunda hangi değerlerin yazdırılacağını tahmin etmeye çalışın

x = 2
def f():
    x = 3
    return x

print(x)      # x = 2 global scope
print(f())    # x = 3 local scope
```

### DEFAULT AND FLEXIBLE ARGUMENTS
1. *Default argument örneği:*
    
    def f(a, b=1):
        """b = 1 is default argument"""
        
1. *Flexible argument örneği:*
  
  <br>def f(*args):
       """ *args kullanarak bir ya da daha fazla argüman alabiliriz."""

    def f(**kwargs)
       """kwargs bir dictionary veri yapısıdır."""
       
<br> Birkaç pratik yapalım.

```python
# default arguments

def f(a, b = 1, c = 2):
    y = a + b + c
    return y

print(f(5))

# peki default arguments olarak belirlenen değişkenlere farklı bir değer atamak istersem?
print(f(5,4,3))

# atarım :)
```

```python
# Flexible arguments: *args
# Fonksiyona değişken sayıda parametre yollamak istersek *args kullanırız.

def f(*args):
    for i in args:
        print(i)
        
f(1)
print("")
f(1,2,3,4)

# Flexible arguments **kwargs dictionary yapısına değişken sayıda parametre göndermek için kullanılır.
def f(**kwargs):
    """ print key and value of dictionary"""
    
    # Eğer aşağıdaki kısmı anlamadıysanız for döngüsü kullanarak dictionary işlemleri yaptığımız bölümü tekrar edin
    for key, value in kwargs.items():
        print(f"Key değeri: {key} , Value değeri: {value}")
        
f(country = 'spain', capital = 'madrid', population = 123456)
```

### ITERATORS - YİNELEYİCİLER
* *iterators*, Python’a iterasyon özelliği kazandıran bir nesnedir. Verileri döndüren bu nesne, her seferinde bir öğe döndürür. Iterators, Pyhton programlama dilinde döngüler, liste, tuple ve generators gibi noktalarda karşımıza çıkar. 
* Python iterators nesnesi, `__iter__()` ve `__next__()` şeklinde iki özel metot içerir. Bu metotlardan bir değer alabilirsek iterator özelliğindeki bir nesne içinde olduğumuzu anlarız. Örneğin Python’da list, tuple, string ve dictionary bu özelliğe sahiptir.
- `__iter__()` -> Bir iteratörün başlatılmasında çağrılır. Bir sonraki adıma geçebilen veya `__next__` yöntemine sahip bir nesne döndürmelidir. Yani geriye `iterable` bir nesne döndürür.
- `__next()` -> **Iterator**, iterable bir obje için bir sonraki değeri döndürür.

```python
# iteration örneği

name = "ronaldo"
itr = iter(name)    # it değişkeni, name stringini alıp iterable olan bir nesneyi alıyor
print(next(itr))    # biri sonraki iteration'u yazdır
print(*itr)         # iterable nesnenin geri kalanını yazdır.

name = "karabük"
itr2 = iter(name)

for i in itr2:
    print(i)
```

# LIST COMPREHENSİON
**Bu tutoria içerisindeki en önemli konulardan birine giriş yapıyoruz.**

* List Comprehension yöntemini, sıklıkla `Data Analysis- Veri Analizi` yapmak için kullanırız.
* List Comprehension yöntemi, *for* döngüsü ile iki üç satırda tanımlayabileceğimiz listeleri tek satırda tanımlama imkanı sunar.
* Örnek olarak, num1 = [1, 2, 3] listesinin içerisindeki değerleri bir arttırarak num2 = [2, 3, 4] haline getirmek istiyoruz. Bu işlemi pek ala bir for döngüsü ile yapabiliriz ancak hem zaman hem de performans verimliliği sağlamak adına *List Comprehension* tekniğini uygulayacağız.


## List Comprehension Örneği

```python
# Klasik yöntem
num1For = [1,2,3]
num2For = [0,0,0]

for index, value in enumerate(num1For):
    num2For[index] = value + 1

print(num2For)
```

```python
# list comprehension örneği

num1 = [1,2,3]
num2 = [i + 1 for i in num1]

print(num2)
```

* `[i + 1 for i in num1]` -> list of comprehension
    
    * i +1: list comprehension sentaksı
    * for i in num1: for söngüsü sentaksı
    * i: iteratör
    * num1: itere edilebilen nesne

```python
# List Comprehension yaparken döngüler ve koşullar kullanmak

num1 = [5,10,15]
num2 = [i**2 
        if i == 10 
        else 
        i-5 
        if i < 7 
        else
        i+5 
        for i in num1]

print(num2)
```

**Yukarıdaki işlemleri şu şekilde açıklayabiliriz:**
* Öncelikle *for i in num1* döngüsünü dikkate alarak başlayacağız;
    * Eğer i == 10 True ise;
         * `for in num1` döngüsünü `i**2` için çalıştırırız
    * Eğer i < 7 True ise;
         * `for in num1` döngüsünü `i-5` için çalıştırırız
    * Eğer i == 10 False ve i < 7 False ise;
         * `for in num1` döngüsünü `i+5` için çalıştırırız
         

```python
# Şimdi de pokemon.csv dosyamız üzerinde List Comprehension yapalım.
# Pokemonların "Speed" feature'unu göz önünde bulundurarak "yavaş" ya da "hızlı" olup olmadılarına bakalım.
# Karşılaştırma değerimizi ortalama hız olacak şekilde belirliyoruz.

threshold = sum(data["Speed"]) / len(data["Speed"])    # Ortalama Hız hesabı
print(f"Threshold Value = {threshold}")

data["SpeedLevel"] = ["High" 
                       if i > threshold 
                       else 
                       "Low" 
                       for i in data["Speed"]]

data.loc[:10,["SpeedLevel","Speed"]]             # loc metodunu ilerleyen zamanlarda öğreneceğiz
```

### EXPLORATORY DATA ANALYSIS - KEŞİFSEL VERİ ANALİZİ
* `value_counts() metodu`: Verinin keşfi sırasında kullanılan metottur. Sütundaki(features), NaN olmayan her bir unique değerin kaç kez kullanıldığını gösteren bir seri döndürür. Bu seri default olarak azalan sırada(descending) gösterilir. Aslında features değerlerinin frequency bilgisine ulaşıyoruz. Bu bilgi üzerinden yorumlamalar yaparak hatalı veri bulma şansımızı arttırabiliriz.


* `Outlier(Aykırı Değer)`: Halihazırda yapılmış olan gözlemlerden kayda değer derecede uzak olan, gözleme aykırı veya uç değerlere *outliers* denir. Aykırı gözlemler, veri setinin geri kalanından farklı davranır ve bu nedenle dikkat çeker. Veri içerisindeki değerlerin dağımı ile kıyaslandığında, gözle görülür derecede yüksek ya da düşük değerleri outlier olarak değerlendirebiliriz.


* Diyelim ki Q3 = %75 ve Q1 = %25. Buna göre;
    * **Q3 + 1.5(Q3-Q1) < OUTLIER < Q1 - 1.5(Q3-Q1)**
        * **Q3 - Q1 = IQR(Interquartile Range - Çeyrekler Açıklığı)**
    * Outlier değerleri Q1 - 1.5(Q3 - Q1) değerinde küçük ve Q3 + 1.5( Q3 - Q1) değerinden büyüktür.


* Ayrıca `describe()` metodunu kullanacağız. describe() metodu aşağıdakileri içerir:
    * count: veri içerisindeki girdi sayısı
    * mean: girdilerin ortalaması
    * std: standart deviation(standart sapma)
    * min: minimum girdi
    * 25%: ilk yüzdelik dilim(firs quantile)
    * 50%: ortanca değer ya da ikinci yüzdelik dilim(second quantile or median)
    * 75%: üçüncü yüzdelik dilim(third quantile)
    * max: maksimum girdi


* `Quartile(Çeyrek)` Nedir?
    * 1, 4, 5, 6, 8, 9, 11, 12, 13, 14, 15, 16, 17
    * Median(medyan), sayı dizisi içerisindeki ortanca değerdir. Bu senaryoda median = 11 
    * `Lower Quartile(İlk Çeyrek/Birinci Kartil)`: Serinin başındaki ile serinin ortasındaki sayı arasında kalan sayıların medyanına denir. Bu senaryoda 1-11 arasındaki değerlerin medyanı olan 6 lower quartile'dir.
    * `Upper Quartile(Üst Çeyrek/Üst Kartil)`: Serinin ortasındaki sayı ile serinin sonundaki sayı arasında kalan sayıların medyanına denir. Bu senaryoda 11-17 arasındaki değerlerin medyanı olan 14 upper quartile'dir.


* `Quantile ve Quartile İlişkisi`
    * **Lower Quartile = Q1 = %25 Quantile**
    * **Upper Quartile = Q3 = %75 Quantile**

### RESAMPLING PANDAS TIME SERIES
* Resampling: Popülasyon hakkında çıkarımlar yapabilmek için bir veri setinden yerine koyarak (sampling with replacement) tekrar tekrar aynı büyüklükte örneklemler oluşturmaya yeniden örnekleme (resampling) denir. Bu yöntemler sayesinde örneklemlerin belirsizliği ve değişkenliği ile ilgili daha iyi fikir elde edilebilir.
    * Resampling yaparken "Ay" ve "Yıl" bazı alınabilir. Ay için resample("M"), yıl için ise resample("A") kullanılır.
* Downsampling: date-time object yapısındaki dataframe satırlarını günlük değişimden haftalık değişime doğru biraz daha seyrek frekansa indirmemizi sağlayan yöntem.
* Upsampling: date-time object yapısındaki dataframe satırlarının sıklığı arttırmamızı sağlayan yöntem. Örneğin günlük değişimden saatlik değişime çevrim gibi.
* Interpolate: 'Linear', "time" veya 'index' gibi farklı teknikleri ya da verileri baz alarak NaN değerleri düzenlemek ve eksik verileri doldurmak için kullandığımuz yöntem. Bahsedilen paratmetreler ile birlikte başka birçok çok parametre baz alınabilir. Bu yönteme dair bilgi almak için:
    * https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.interpolate.html


```python
# data2 data frame'i üzerinde yıl verisini baz alarak resampling yapalım.
# Burada yıllara göre feature'ların ortalama değerlerini inceliyoruz.
data2.resample("A").mean()
```

```python
# Şimdi de aylara göre resampling yapalım
data2.resample("M").mean()
# Görüldüğü üzere yer yer NaN veriler mevcut, bunun nedeni ise data2'de bazı aylara ait verilerin mevcut olmamasıdır.
```

```python
# Gerçek hayatın gerçek verileri ile çalışırken boş verileri doldurmamız gerekebilir.
# Bunu gerçekleştirmek için "interpolate" yöntemini kullanırız.
# Interpolate yöntemi ile yalnızca nümerik veriler doldurulabilir. 
# İlk değerden itibaren interpolate işlemini başlatalım.
# Burada interpolate yöntemini linear parametresi kullanarak gerçekleştiriyoruz
data2.resample("MS").first().interpolate("ffill")
```

```python
# İlk değerden itibaren değil de ortalamasını alarak linear parametresi ile birlikte interpolate işlemi yapabiliriz.
data2.resample("M").mean().interpolate("linear")
```
