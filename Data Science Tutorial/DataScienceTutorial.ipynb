### MATPLOTLIB
Matplotlib, verinin bir çizimini(plot) oluşturarak görselleştirmemizi sağlayan kütüphanedir. Oluşturması en basit ve en çok kullanılan diyagramlar ise line(çizgi), scatter(serpilme) ve histogram plot'larıdır.
* Grafiğin x ekseninde "zaman" değişkeni mevcutsa Line Plot kullanmak mantıklıdır.
* İki değişken/feature arasında korelasyon(correlation) mevcutsa Scatter Plot kullanmak mantıklıdır.
* Sayısal verilerin dağılımını görmek istiyorsak Histogram kullanmak mantıklıdır.
* Customization(Özelleştirme): Renkler, etiketler(labels), çizgi kalınlığı, başlık, saydamlık, ızgara(grid), şekil boyutu(figsize), eksen çizgileri ve çizgi sitili(linestyle)

```python
# Line Plot(Çizgi Grafiği)
# color = renk, label = etiket, linewidth = çizgi kalınlığı, alpha = saydamlık, grid = grid, linestyle = sytle of line

data.Speed.plot(kind = 'line', color = 'g', label = 'Speed', linewidth=1, alpha = 0.5, grid = True, linestyle = ':')
data.Defense.plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     # legend = eksen etiketlerini ve konumunu belirle
plt.xlabel('x axis')              # label = etiket ismi
plt.ylabel('y axis')
plt.title('Line Plot')            # title = grafiğin başlığı
plt.show()
```

```python
# Scatter Plot(Serpilme Grafiği)
# İki feature arasındaki positive correlation ve negative correlation değişimlerini inceleyeceğiz
# Feature'lar: x = Attack, y = Defense

data.plot(kind='scatter', x='Attack', y='Defense', alpha = 0.5, color = 'red')
plt.xlabel('Attack')
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot')
plt.show()
```

### DICTIONARY(SÖZLÜK)
Neden dictionary veri yapısına ihtiyaç duyarız?
* Çünkü dictionary, tıpkı küçük bir dataframe gibi içerisinde 'key' ve 'values' değerlerini bulundurur.
* Dictionary, list veri yapısından daha hızlı çalışır.
<br>
'key' ve 'value' değerleri nedir? Örnek:
* dictionary = {'Spain' : 'Madrid'}
* Key >>> Spain
* Values >>> Madrid
<br>

**İşte bu kadar basit**

Şimdi de, keys(), values(), güncelleme(update), ekleme(add), kontrol etme(chech), key silme(remove key), tüm girdileri silme(remove all entries) ve dictionary silme(remove dictinary) gibi metot ve özellikler ile alakalı uygulamalar yapalım.

```python
# Bir dictionary yaratıp key ve values değerlerini inceleyim

dictInstance = {
    'Spain' : 'Madrid',
    'USA' : 'Vegas'
}

print(dictInstance.keys())
print(dictInstance.values())

# Görüldüğü üzere dictInstance adlı dictionary yapısının;

    # Key değerleri >> Spain ve USA
    # Values değerleri >> Madrid ve Vegas
```

```python
# Dictionary key'leri; tıpkı string, boolen, float, integer ya da tuple gibi immutable nesnelerdir.
# List veri yapısı mutable'dır
# Key değerleri unique yani eşşizdir.

print(dictInstance)                    # dictInstance dictionary'sini yazdır ve ilk halini incele.

dictInstance['Spain'] = "Barcelona"    # Halihazırda var olan key'in values değişkenini güncelledik.
print(dictInstance)

dictInstance['France'] = "Paris"       # Yeni bir key-values ikilisini dictInstance yapımıza ekledik.
print(dictInstance)

del dictInstance['Spain']              # Spain key'ini, values değişkeni ile birlikte sildik.
print(dictInstance)

print('France' in dictInstance)        # France key'inin dictInstance yapısı içerisinde olup olmadığını kontrol et.

dictInstance.clear()                     # dictInstance içerisindeki tüm girdileri sil.
print(dictInstance)
```

```python
# Boş bir dictionary yapısının bellekte yer kaplaması mantıklı değil.
# Bu nedenle dictInstance'ı tamamen silebiliriz.
# del dictInstance ifadesi dictionary yapısını silmemizi sağlar.
# Silinen bir dictionary yapısını yazdırmak hata vereceği için print() satırını yorum satırı yaptık.

del dictInstance
# print(dictInstance)      
```

### WHILE AND FOR LOOPS - WHILE VE FOR DÖNGÜLERİ
* `while` ve `for` döngülerini en basit şekilde öğreneceğiz.

```python
# (i != 5) şartı True olduğu sürece while döngüsü içerisinde kal

i = 0
while i != 5:
    print('i is: ', i)
    i +=1
print(i,' is equal to 5')
```

```python
print("-------For List-------")

# for döngüsü örneği

lis = [1,2,3,4,5]
for i in lis:
    print('i is: ',i)
print('')

print("-------For Enumerate-------")

# enumerate -> liste yapısı içerisindeki değerleri ve o değerlerin listedeki indekslerini yazdırmamızı sağlar
# index : value = 0:1, 1:2, 2:3, 3:4, 4:5

for index, value in enumerate(lis):
    print(index," : ",value)
print('')   

print("-------For Dictionary-------")

# Dictionary yapısında döngüleri inceleyelim
# Dictionary'nin key ve values değerlerine erişebilmek için for döngüsü oluşturabiliriz.
# Bunu yapabilmek için dictionary veri yapısına ait olan items() metodunu kullanabiliriz.

dictionary = {
    'Spain':'Madrid',
    'France':'Paris'
}

for key, value in dictionary.items():
    print(key," : ",value)
print('')

print("-------For Pandas-------")

# Pandas verilerine erişim sağlamak için for döngüsü kullanabiliriz.
# Attack feature'unun ilk indeksindeki değeri -> [0:3] 
# iterrows() metodu ile, belirtilen indeksi ve o indeksteki değeri alıp index ve value değişkenlerine atıyoruz.
# Ardından for döngüsü ile yazdırıyoruz
for index, value in data[['Attack']][0:3].iterrows():
    print(index," : ",value)
```

Peki, fonksiyonlar ile ilgili bilmemiz gereken kavramlar nelerdir?
* docstrings: ait olduğu fonksiyonun işlevini özetleyen dökümanlardır. Örneğin:
<br>for f():
    <br>"""This is docstring for documentation of function f"""


* tuple: birbiri ardına sıralanmış Python objelerini tutan veri yapısı.
1. Tuple içerisindeki değerler değiştirilemez çünkü tuple immutable'dır.
2. tuple içerisindeki sıralı objeler parantez içerisinde belirtilir -> tupleInstance = (a,b,c)
3. tuple içerisinde depolanan verileri birden fazla değişkene aktarabiliriz ve bu işleme `unpacking` denir

```python
# Öğrendiklerimizi uygulayalım

def tuple_ex():
    """ return defined t tuple"""  # docstring
    t = (1,2,3)
    return t

a,b,c = tuple_ex()     # unpacking
print(a,b,c)
```

### SCOPE - KAPSAM / FALİYET ALANI
Peki, scope hakkında bilmemiz gereken kavramlar nelerdir?
* local: bir fonksiyon içerisinde tanımlanan değişkenlerin türü
* enclosed: eğer yerelde tanımlı değilse kapsayan fonksiyon içinde yer alan değer kullanılır
* global: main body(ana gövde) içerisinde tanımlanan değişkenlerin türü
* built in scope: Bizim tanımlamadığımız, Python ile birlikte gelen fonksiyon ya da değişkenlerdir. Örn: print(), len()
<br><br>Lets make some basic examples

```python
# print() fonksiyonu sonucunda hangi değerlerin yazdırılacağını tahmin etmeye çalışın

x = 2
def f():
    x = 3
    return x

print(x)      # x = 2 global scope
print(f())    # x = 3 local scope
```

```python
# Peki kullanmak istediğimiz değişken local scope içerisinde tanımlı değilse ne olur?

x = 5              # global değişken

def f():
    y = 2*x        # fonksiyonda, yani local scope içerisinde tanımlı bir x değişkeni yok
    return y

print(f())         # global scope içerisinde tanımlanan x değişkeni kullanılır.

# Yukarıdaki script'i inceleyelim;
# f() fonksiyonu içerisinde x değişkeni aranır, 
# Eğer yoksa global scope içerisinde bir x değişkeni olup olmadığına bakılır.
# Eğer global scope içerisinde de bir x değişkeni yoksa built in scope içerisinde bir x değişkeni aranır.
```

### NESTED FUNCTION - İÇ İÇE FONKSİYON
* LEGB kuralına göre değişken arama sırası:
    * Local scope
    * Enclosed function
    * Global scope
    * Built in scope

```python
# İç içe fonksiyon örneği

def square():
    """ return square of value """
    
    def add():
        """ add two local variable """
        x = 2
        y = 3
        z = x + y
        return z
    
    return add()**2

print(square())    
```

### DEFAULT AND FLEXIBLE ARGUMENTS
1. Default argument örneği:
    
    def f(a, b=1):
        """b = 1 is default argument"""
        
1. Flexible argument example:
    def f(*args):
       """ *args can be one or more"""

    def f(**kwargs)
       """kwargs is a dictionary"""
       
<br><br> Birkaç pratik yapalım.

```python
# default arguments

def f(a, b = 1, c = 2):
    y = a + b + c
    return y

print(f(5))
# peki default arguments olarak belirlenen değişkenlere farklı bir değer atamak istersem?
print(f(5,4,3))
# atarım :)
```

```python
# flexible arguments: *args
# fonksiyona değişken sayıda parametre yollamak istersem *args kullanırım

def f(*args):
    for i in args:
        print(i)
        
f(1)
print("")
f(1,2,3,4)

# flexible arguments **kwargs dictionary yapısına değişken sayıda parametre göndermek için kullanılır.
def f(**kwargs):
    """ print key and value of dictionary"""
    
    # Eğer aşağıdaki kısmı anlamadıysanız for döngüsü kullanarak dictionary işlemleri yaptığımız bölümü tekrar edin
    for key, value in kwargs.items():
        print(key, " ", value)
        
f(country = 'spain', capital = 'madrid', population = 123456)
```

### ITERATORS - YİNELEYİCİLER
* Python’a iterasyon özelliği kazandıran bir nesnedir. Verileri döndüren nesne her seferinde bir öğe döndürür. Iterators, Pyhton programlama dilinde döngüler, liste, tuple ve generators gibi noktalarda karşımıza çıkar. 
* Python iterators nesnesi, `__iter__()` ve `__next__()` iki özel metot içerir. Bu metotlardan bir değer alabilirsek iterator özelliğindeki bir nesne içinde olduğumuzu gösterir. Örneğin Python’da list, tuple, string ve dictionary bu özelliğe sahiptir.
- `__iter__()` -> Bir iteratörün başlatılmasında çağrılır. Bir sonraki veya `__next__` yöntemine sahip bir nesne döndürmelidir. Yani geriye `iterable` bir nesne döndürür.
- `__next()` -> **Iterator**, iterable bir obje için bir sonraki değeri döndürür.

```python
# iteration örneği

name = "ronaldo"
it = iter(name)    # it değişkeni, name stringini alıp iterable olan bir nesneyi alıyor
print(next(it))    # biri sonraki iteration'u yazdır
print(*it)         # iterable nesnenin geri kalanını yazdır.

name = "karabük"
iterr = iter(name)

for i in iterr:
    print(i)
```

# LIST COMPREHENSİON
**Bu tutoria içerisindeki en önemli konulardan birine giriş yapıyoruz**

* List Comprehension yöntemini, sıklıkla `Data Analysis- Veri Analizi` yapmak için kullanırız.
* List Comprehension: for döngüsü ile iki üç satırda tanımlayabileceğimiz listeleri tek satırda tanımlama imkanı sunar.
* Örnek olarak, num1 = [1, 2, 3] listesinin içerisindeki değerleri bir arttırarak num2 = [2, 3, 4] haline getirmek istiyoruz. Bu işlemi pek ala bir for döngüsü ile yapabiliriz ancak hem zaman hem de performans verimliliği sağlamak adına List Comprehension tekniğini uygulayacağız.


## Example of List Comprehension

```python
# klasik yöntem
num1_for = [1,2,3]
num2_for = [0, 0,0]

for index, value in enumerate(num1_for):
    num2_for[index] = value + 1

print(num2_for)

# list comprehension örneği

num1 = [1,2,3]
num2 = [i + 1 for i in num1]

print(num2)
```

[i + 1 for i in num1 ]: list of comprehension
<br> i +1: list comprehension syntax
<br> for i in num1: for loop syntax
<br> i: iterator
<br> num1: iterable object

```python
# Conditionals on iterable
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```
